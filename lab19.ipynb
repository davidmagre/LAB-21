```python
#Concatenate the three dataframes
```

```python
column_names = file1
data = pd.DataFrame(columns=column_names)
data = pd.concat([file1, file2, file3], axis=0)
data
```

```python
#Understand the meaning of all columns

#customer = codigo de cada cliente
#state = estado de procedencia de cada cliente
#gender = genero de cada cliente
#education = nivel de educacion de cada cliente
#clv = customer lifetime value, valor que aporta cada cliente a una empresa tanto actual como potencial
#income = ingresos por parte de cada cliente
#mpa = monthly premium auto, monto de dinero que pagas de forma mensual a la aseguradora
#nop = number of open complaints, numero de casos abiertos con la aseguradora actualmente
#policy_type = type of car policy
#vehicle_class = type of car
#tca = total claim amount, monto de dinero pedido por el seguro en total de disputas
```

```python
#Perform the data cleaning operations mentioned so far in class
```

```python
#Delete the column education and the number of open complaints from the dataframe.
```

```python
columns_to_drop = ['education', 'nop']
data = data.drop(columns=columns_to_drop)
```

```python
data.head(1)
```

```python
#Correct the values in the column customer lifetime value. 
#They are given as a percent, so multiply them by 100 and change `dtype` to `numerical` type.
```

```python
data['clv'] =  pd.to_numeric(data['clv'], errors='coerce')*100
```

```python
data._get_numeric_data()
```

```python
#Check for duplicate rows in the data and remove if any.
```

```python
data = data.drop_duplicates()
```

```python
#Filter out the data for customers who have an income of 0 or less.
```

```python
data[(data['income']<=0)]
```

```python
#Check and deal with `NaN` values.
```

```python
round(data.isna().sum()/len(data),4)*100
```

```python
data = data.drop(['clv'], axis=1)
```
